### Pertanyaan 2: Apakah dalam satu tahun terakhir terjadi peningkatan atau penurunan dalam transaksi pelanggan?

```python
monthly_transactions_df = all_data.resample(rule='M', on='order_approved_at').agg({
    "payment_value":"sum"
})
monthly_transactions_df.index = monthly_transactions_df.index.strftime('%B') #mengubah format order_approved_at menjadi Tahun-Bulan
monthly_transactions_df = monthly_transactions_df.reset_index()
monthly_transactions_df.rename(columns={
    "payment_value":"total_transactions"
}, inplace=True)
monthly_transactions_df.head()
```

```python
monthly_transactions_df = monthly_transactions_df.sort_values('total_transactions').drop_duplicates('order_approved_at', keep='last')
monthly_transactions_df.head()
```

```python
monthly_transactions_df.sort_values(by='total_transactions')
```

```python
month_mapping = {
    "January": 1,
    "February": 2,
    "March": 3,
    "April": 4,
    "May": 5,
    "June": 6,
    "July": 7,
    "August": 8,
    "September": 9,
    "October": 10,
    "November": 11,
    "December": 12
}

monthly_transactions_df["month_numeric"] = monthly_transactions_df["order_approved_at"].map(month_mapping)
monthly_transactions_df = monthly_transactions_df.sort_values("month_numeric")
monthly_transactions_df = monthly_transactions_df.drop("month_numeric", axis=1)
```

```python
plt.figure(figsize=(10, 5))
plt.plot(
    monthly_transactions_df["order_approved_at"],
    monthly_transactions_df["total_transactions"],
    marker='o',
    linewidth=2,
    color="#CD5C5C")

plt.title("Total Transaksi pelanggan Perbulan dalam 1 Tahun Terakhir", loc="center", fontsize=20)
plt.xlabel("Bulan", fontsize=12)
plt.ylabel("Total Transaksi", fontsize=12)
plt.xticks(fontsize=10, rotation=25)
plt.yticks(fontsize=10)
plt.grid(True)
plt.show()
```

###Pertanyaan 3: Bagaimana perkembangan penjualan produk dalam satu tahun terakhir? Apakah penjualan meningkat atau menurun?

```python
monthly_sells_df = all_data.resample(rule='M', on='order_approved_at').agg({
    "order_id": "nunique",
})
monthly_sells_df.index = monthly_sells_df.index.strftime('%B')
monthly_sells_df = monthly_sells_df.reset_index()

monthly_sells_df.rename(columns={
    "order_id": "order_count",
}, inplace=True)
monthly_sells_df.head()
```

```python
monthly_sells_df = monthly_sells_df.sort_values('order_count').drop_duplicates('order_approved_at', keep='last')
monthly_sells_df.head()
```

```python
monthly_sells_df.sort_values(by='order_count')
```

```python
monthly_sells_df["month_numeric"] = monthly_sells_df["order_approved_at"].map(month_mapping)
monthly_sells_df = monthly_sells_df.sort_values("month_numeric")
monthly_sells_df = monthly_sells_df.drop("month_numeric", axis=1)
```

```python
plt.figure(figsize=(10, 5))
plt.plot(
    monthly_sells_df["order_approved_at"],
    monthly_sells_df["order_count"],
    marker='o',
    linewidth=2,
    color="#CD5C5C"
)
plt.title("Jumlah Penjualan Produk Perbulan dalam 1 Tahun" , loc="center", fontsize=20)
plt.xlabel("Bulan", fontsize=12)
plt.ylabel("Jumlah Penjualan", fontsize=12)
plt.xticks(fontsize=10, rotation=25)
plt.yticks(fontsize=10)
plt.grid(True)
plt.show()
```
